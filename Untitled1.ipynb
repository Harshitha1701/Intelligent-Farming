```python
import pandas as pd
path="C:\\Users\\SANGEETHA T SHENOY\\Desktop\\SMART FARMING\\SmartCrop-Dataset.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
print(data.shape)
data.head(20)
```

```python
data.isnull().sum()
```

```python
data.rename(columns = {'label':'Crop'}, inplace= True)
data
```

```python
inputs=data.drop('Crop',axis=1)
inputs
```

```python
output=data.Crop
output
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(inputs,output)
res=model.predict([[90,42,43,20.88,82.00,6.50,202.94]])
print("OUTCOME:",res)
acc=model.score(inputs,output)
print("ACCURACY:",acc*100)
```

```python
import pandas as pd
path1="C:\\Users\\SANGEETHA T SHENOY\\Desktop\\SMART FARMING\\fertilizer.csv"
data1=pd.read_csv(path1)
print(data1)
print(data1.info())
print(data1.shape)
data1.head(20)
```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
le_Crop=LabelEncoder()
data1['Crop']=le_Crop.fit_transform(data1['Crop'])
data1
```

```python
inputs1=data1.drop('Fertilizer_Name',axis=1)
inputs1
```

```python
output1=data1.Fertilizer_Name
output1
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression
models=LogisticRegression()
models.fit(inputs1,output1)
res=models.predict([[20,80,40,40,5.5]])
print("OUTCOME:",res)
acc=models.score(inputs1,output1)
print("ACCURACY:",acc*100)
```

```python
import matplotlib.pyplot as plt
x=data1["Crop"]
y=data1["Fertilizer_Name"]
plt.scatter(x,y)
plt.show()
```
